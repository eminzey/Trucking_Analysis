```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Define truck models, types, and their respective costs
truck_models = [
    ('Tesla Semi 24', 'Electric', 0.110273, 0.089831),
    ('BYD Semi 24', 'Electric', 0.219525, 0.125104),
    ('Kenworth T680 24', 'Diesel', 0.494483, 0.357259),
    ('Freightliner Cascadia 24', 'Diesel', 0.535878, 0.361720),
    ('International LT 24', 'Diesel', 0.598540, 0.380783)
]

# Set a fixed mileage for all trucks
distance_miles = 300

# Create data dictionary
data = {
    'truck_id': [f'TRK{i:03d}' for i in range(1, 1001)],  # Generate 1000 truck IDs
    'truck_model': np.random.choice([model[0] for model in truck_models], size=1000)
}

# Add truck type, fuel cost, maintenance cost, and total cost per mile based on the truck model
data['truck_type'] = [
    next(model[1] for model in truck_models if model[0] == truck_model)
    for truck_model in data['truck_model']
]
data['distance_miles'] = [distance_miles] * 1000
data['fuel_cost_per_mile'] = [
    next(model[2] for model in truck_models if model[0] == truck_model)
    for truck_model in data['truck_model']
]
data['maintenance_cost_per_mile'] = [
    next(model[3] for model in truck_models if model[0] == truck_model)
    for truck_model in data['truck_model']
]

# Calculate total operational cost per mile
data['total_cost_per_mile'] = [
    fuel_cost + maintenance_cost
    for fuel_cost, maintenance_cost in zip(data['fuel_cost_per_mile'], data['maintenance_cost_per_mile'])
]

# Create a DataFrame from the generated data
trucking_data = pd.DataFrame(data)

# Calculate total cost for each truck entry based on the distance
trucking_data['total_cost'] = trucking_data['distance_miles'] * trucking_data['total_cost_per_mile']

# Show first few rows of generated data
print(trucking_data.head())

# Group data by truck model and calculate average cost per mile
avg_cost_per_mile = trucking_data.groupby('truck_model')['total_cost_per_mile'].mean()
print(avg_cost_per_mile)

# Plot comparison of average operational cost per mile
plt.figure(figsize=(15, 10))

# Subplot 1: Total cost per mile
plt.bar(avg_cost_per_mile.index, avg_cost_per_mile.values, color=['lightgreen', 'lightgreen', 'skyblue', 'skyblue', 'skyblue'])
plt.xlabel('Truck Model')
plt.ylabel('Total Cost per Mile (USD)')
plt.title('Total Cost per Mile by Truck Model')

plt.tight_layout()
plt.show()
```
